```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import random
import time

def get_motor(basic_info):
    tables = []
    motors = []
    mileages = []
    data = [motors, mileages]
    for item in basic_info:
        for i in item:
            table = i.find_all("table", attrs={"class": "used-specs-table"})
            if table:
                tables.append(table[0])
    for table in tables:
        motor = table.find("td", attrs={"class": "car-ad-cc"})
        mileage = table.find("td", attrs={"class": "car-ad-km"})
        if motor:
            motors.append(motor.string)
        else:
            motors.append(None)
        if mileage:
            mileages.append(mileage.string)
        else:
            mileages.append(None)
    return data

def get_basic_info(content_list):
    basic_info = []
    for content in content_list:
        info = content.find_all('div', class_='some-class')  # Adjust this according to the actual HTML structure
        basic_info.append(info)
    return basic_info

def get_names(basic_info):
    names = []
    for info in basic_info:
        name = info[0].find('span', class_='car-name')  # Adjust this according to the actual HTML structure
        names.append(name.string if name else None)
    return names

def get_prices(basic_info):
    prices = []
    for info in basic_info:
        price = info[0].find('span', class_='car-price')  # Adjust this according to the actual HTML structure
        prices.append(price.string if price else None)
    return prices

def get_years(basic_info):
    years = []
    for info in basic_info:
        year = info[0].find('span', class_='car-year')  # Adjust this according to the actual HTML structure
        years.append(year.string if year else None)
    return years

page_number = 1
names = []
prices = []
years = []
motors = []
mileages = []


headers = {'User-Agent': 'Mozilla/5.0'}

for i in range(9):
    base_url = "https://www.cardekho.com/used-cars+in+mumbai?page={}".format(page_number)
    response = requests.get(base_url, headers=headers)
    html_soup = BeautifulSoup(response.text, 'html.parser')
    content_list = html_soup.find_all('div', attrs={'class': 'car-ad sft-ad'})

    if not content_list:
        break

    basic_info = get_basic_info(content_list)
    if not basic_info:
        continue

    names1 = get_names(basic_info)
    prices1 = get_prices(basic_info)
    years1 = get_years(basic_info)
    motors1, mileages1 = get_motor(basic_info)

    names.extend(names1)
    prices.extend(prices1)
    years.extend(years1)
    motors.extend(motors1)
    mileages.extend(mileages1)
    page_number += 1
    time.sleep(random.randint(1, 2))

cols = ["Name", "Year", "Motor", "Mileage (Km)", "Price"]
data = pd.DataFrame({"Name": names, "Year": years, "Motor": motors, "Mileage (Km)": mileages, "Price": prices})[cols]
data["Price"] = data["Price"].replace({'\$ ': ''}, regex=True)
data["Price"] = data["Price"].replace({',': ''}, regex=True)
data["Mileage (Km)"] = data["Mileage (Km)"].replace({' Km': ''}, regex=True)
data[["Mileage (Km)", "Year", "Motor", "Price"]] = data[["Mileage (Km)", "Year", "Motor", "Price"]].apply(pd.to_numeric, errors='coerce')

data.drop_duplicates().to_excel('Car_list.xlsx', engine='openpyxl')

data.head()
```

```python
import requests
from bs4 import BeautifulSoup
import csv

def fetch_webpage(url, headers):
    """Fetch the webpage content."""
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return response.content
    else:
        print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
        return None

def parse_car_data(soup):
    """Parse car data from the BeautifulSoup object."""
    car_list = []
    for car in soup.find_all('div', class_='gsc_col-xs-12 gsc_col-sm-6 gsc_col-md-4 cardColumn'):
        model_name = car.find('h3', class_='title').text if car.find('h3', class_='title') else 'N/A'
        image_tag = car.find('img')
        image = image_tag['src'] if image_tag else 'N/A'
        link_tag = car.find('a')
        link = link_tag['href'] if link_tag else 'N/A'
        price_tag = car.find('div', class_='Price')
        price = price_tag.text.strip() if price_tag else 'N/A'
        details = car.find('div', class_='dotsDetails')
        fuel_type, transmission = 'N/A', 'N/A'
        if details:
            details_list = details.text.split('�')
            if len(details_list) > 1:
                fuel_type = details_list[1].strip()
            if len(details_list) > 2:
                transmission = details_list[2].strip()
        car_list.append({
            'model_name': model_name,
            'image': image,
            'link': link,
            'price': price,
            'fuel_type': fuel_type,
            'transmission': transmission
        })
    return car_list

def write_to_csv(car_list, filename):
    """Write car data to a CSV file."""
    with open(filename, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=['model_name', 'image', 'link', 'price', 'fuel_type', 'transmission'])
        writer.writeheader()
        writer.writerows(car_list)
    print(f"Car data has been written to {filename}")

def main():
    url = "https://www.cardekho.com/used-volvo+cars+in+mumbai"
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }
    html_content = fetch_webpage(url, headers)
    if html_content:
        soup = BeautifulSoup(html_content, 'html.parser')
        car_list = parse_car_data(soup)
        write_to_csv(car_list, 'used_volvo_cars_in_mumbai.csv')

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
import re

# Load the CSV file into a DataFrame
df = pd.read_csv('used_volvo_cars_in_mumbai.csv')

# Display the first few rows of the DataFrame before cleaning
print("Before cleaning:")
print(df.head())

# 1. Remove duplicates
df.drop_duplicates(inplace=True)

# 2. Handle missing values
df.fillna({
    'model_name': 'Unknown',
    'image': 'No Image',
    'link': 'No Link',
    'price': 'Not Available',
    'fuel_type': 'Unknown',
    'transmission': 'Unknown'
}, inplace=True)

# 3. Standardize text data
df['model_name'] = df['model_name'].str.strip().str.title()
df['fuel_type'] = df['fuel_type'].str.strip().str.title()
df['transmission'] = df['transmission'].str.strip().str.title()

# 4. Clean price column
def clean_price(price):
    try:
        # Remove '₹' and 'Compare', replace 'Lakh' with * 100000, and convert to float
        price = price.replace('₹', '').replace('Compare', '').strip()
        if 'Lakh' in price:
            price = price.replace('Lakh', '').strip()
            price = float(price) * 100000
        else:
            price = float(price)
        return price
    except:
        return np.nan

df['price'] = df['price'].apply(clean_price)

# 5. Remove unwanted characters or whitespaces
df['link'] = df['link'].str.strip()
df['image'] = df['image'].str.strip()

# 6. Separate year from model_name
def extract_year(model_name):
    match = re.match(r'(\d{4})\s+(.*)', model_name)
    if match:
        year = match.group(1)
        model = match.group(2)
        return pd.Series([year, model])
    else:
        return pd.Series([np.nan, model_name])

df[['year', 'model_name']] = df['model_name'].apply(extract_year)

# Display the first few rows of the DataFrame after cleaning
print("\nAfter cleaning:")
print(df.head())

# Optionally, save the cleaned DataFrame to a new CSV file
df.to_csv('cleaned_used_volvo_cars_in_mumbai.csv', index=False)
```

```python
# @title year vs price

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['year']
  ys = series['price']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('year', ascending=True)
for i, (series_name, series) in enumerate(df_sorted.groupby('image')):
  _plot_series(series, series_name, i)
  fig.legend(title='image', bbox_to_anchor=(1, 1), loc='upper left')
sns.despine(fig=fig, ax=ax)
plt.xlabel('year')
_ = plt.ylabel('price')
```
